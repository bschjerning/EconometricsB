```python
# Illustrating convergence in probability
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
mu=1
sigma=2
def f_N(z,N):
    return norm.pdf(z,mu,sigma/np.sqrt(N))
h=2
zgrid=np.linspace(mu-h,mu+h,1000)
plt.plot(zgrid,f_N(zgrid,N=10))
plt.plot(zgrid,f_N(zgrid,N=20))
plt.plot(zgrid,f_N(zgrid,N=100))
ax=plt.gca()
ax.axes.xaxis.set_ticklabels([])
ax.axes.yaxis.set_ticklabels([])
ax.legend(['${N=10}$','${N=20}$','${N=100}$'])
plt.text(1,-.25,r'${\mu}$')
plt.title("PDF of $\overline{Z}_N$ as function of $N$ with $Z_i$ independent $\mathrm{N}(\mu,\sigma^2)$")
plt.savefig('img/conv_in_prob.png',dpi=300)
plt.show()

```
